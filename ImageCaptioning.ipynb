```python
with open('./sample_data/Flickr8k.lemma.token.txt','r') as f:
    token_entries = f.readlines()
```

```python
imagedict = dict()
for token in token_entries:
  imageId , image_descr = token.split('\t')
  image_descr = image_descr.split('\n')[0]
  newImageId, seq = imageId.split('\t')[0].split('#')
  if(newImageId not in imagedict.keys() ):
    imagedict[newImageId] = list()
  imagedict[newImageId].append(image_descr)
  # print(imageId, image_descr)
```
